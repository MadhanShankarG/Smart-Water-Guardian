```python
import numpy as np
import pandas as pd
import joblib
import tensorflow as tf

from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("pond_dataset.csv")

print(df.head())
print(df.columns)
```

```python
X = df[["pH", "TDS", "temperature"]]
```

```python
def create_label(row):
    if (
        6.5 <= row["pH"] <= 8.5 and
        row["TDS"] <= 500 and
        24 <= row["temperature"] <= 27
    ):
        return 1   # SAFE
    return 0       # UNSAFE

y = df.apply(create_label, axis=1)

print("Safe count:", sum(y))
print("Unsafe count:", len(y)-sum(y))
```
